# Imports

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

# Reading the data

```python
Data = pd.read_csv("covid_day_wise.csv")
```

# Data preparation
## Check  values

```python
Data.isna().sum()
```

```python
Data.info()
```

## Convert Date to datetime data type

```python
Data["Date"] = pd.to_datetime(Data["Date"])
```

## Set Date as index

```python
Data.set_index('Date', inplace=True)
```

## Sort the new index

```python
Data.sort_index()
```


<p>Deaths/Confirmed is the total fatality rate</p>

```python
Data['total Deaths'] = Data.apply(lambda row: row['Deaths'] / row['Confirmed'], axis=1)
Data.plot(kind='line', y='total Deaths',color='Purple')
```


<p>Revocered/Confirmed is the total survival rate</p>

```python
Data['total survival'] = Data.apply(lambda row: row['Recovered'] / row['Confirmed'], axis=1)
Data.plot(kind='line', y='total survival',color='green')
```
